```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
```

```python
iris = datasets.load_iris()
iris.keys()
```

```python
from sklearn.preprocessing import StandardScaler

std_scaler = StandardScaler()
scaled_data = std_scaler.fit_transform(iris['data'])
```

```python
pca = decomposition.PCA(n_components=3)
```

```python
transformed_data = pca.fit_transform(scaled_data)
transformed_data
```

```python
iris['target']
```

```python
iris['target_names']
```

```python
setosa_index = np.where(iris['target'] == 0)[0]
versicolor_index = np.where(iris['target'] == 1)[0]
virginica_index = np.where(iris['target'] == 2)[0]
```

```python
ax = plt.figure(figsize=(20,10)).add_subplot(projection='3d')

x = transformed_data[setosa_index][:, 0]
y = transformed_data[setosa_index][:, 1]
z = transformed_data[setosa_index][:, 2]

ax.scatter(x, y, z, zdir='z', c='r', label='Setosa')


x = transformed_data[versicolor_index][:, 0]
y = transformed_data[versicolor_index][:, 1]
z = transformed_data[versicolor_index][:, 2]

ax.scatter(x, y, z, zdir='z', c='b', label='Versicolor')


x = transformed_data[virginica_index][:, 0]
y = transformed_data[virginica_index][:, 1]
z = transformed_data[virginica_index][:, 2]

ax.scatter(x, y, z, zdir='z', c='g', label='Virginica')

plt.title('Iris Dataset')
plt.legend()

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```
